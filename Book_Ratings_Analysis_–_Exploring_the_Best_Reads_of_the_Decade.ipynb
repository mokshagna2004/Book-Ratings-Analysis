```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load dataset
file_path = "/content/books_of_the_decade.csv"
dataset = pd.read_csv(file_path, encoding='ISO-8859-1')

# Display first few rows
print("Dataset Overview:")
print(dataset.head())
```

```python
# Checking for missing values
print("\nMissing Values:")
print(dataset.isnull().sum())

# Print actual column names
print("Column Names in Dataset:")
print(dataset.columns)


```

```python
# Convert rating and votes to numeric
dataset['Rating'] = pd.to_numeric(dataset['Rating'], errors='coerce')
dataset['Number of Votes'] = pd.to_numeric(dataset['Number of Votes'], errors='coerce')

# Verify the changes
print("\nUpdated Data Types:")
print(dataset.dtypes)
```

```python
# Top 10 Highest-Rated Books
top_rated = dataset.sort_values(by='Rating', ascending=False).head(10)

print("\nTop 10 Highest-Rated Books:")
print(top_rated[['Book Name', 'Author', 'Rating']])
```

```python
# Most Popular Authors (Authors with the most books in the dataset)
popular_authors = dataset['Author'].value_counts().head(10)

print("\nMost Popular Authors:")
print(popular_authors)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check if dataset is correctly loaded
print(dataset.head())  # Ensure correct dataframe reference

# Plot Rating Distribution
plt.figure(figsize=(10, 5))
sns.histplot(dataset['Rating'], bins=20, kde=True, color="blue")

# Labels and title
plt.title("Distribution of Book Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

```python
# Top 5 authors with most books
top_authors = dataset["Author"].value_counts().head(5).index

# Filter dataset for only those authors
filtered_data = dataset[dataset["Author"].isin(top_authors)]

# Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x="Author", y="Rating", data=filtered_data, palette="coolwarm")

# Labels and title
plt.xlabel("Author")
plt.ylabel("Rating")
plt.title("Rating Distribution of Top 5 Authors")

# Rotate x-axis labels
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
